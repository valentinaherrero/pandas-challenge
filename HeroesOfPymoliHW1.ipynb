```python
import pandas as pd
```

```python
file_to_load = "Resources/purchase_data.csv"
```

```python
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df
```

```python
# The unique method shows every element of the series that appears only once
unique_players = purchase_data_df["SN"].unique()
unique_players
```

```python
# Total number of players
total_player = len(unique_players)
total_player
```

```python
# Display total numer of players in a DF
total_player_df = pd.DataFrame({"Total Players": [total_player]})
total_player_df
```

```python
# Unique items
unique_items = purchase_data_df["Item Name"].unique()
unique_items1 = len(unique_items)
unique_items1
```

```python
# Calculating Average Price
average_price = purchase_data_df["Price"].mean()
average_price
```

```python
# Calculating Number of Purchases
number_purchase_unique = purchase_data_df["Purchase ID"].unique()
number_purchase = len(number_purchase_unique)
number_purchase
```

```python
# Calculating Total Revenue
total_revenue = purchase_data_df["Price"].sum()
total_revenue
```

```python
# Summary data frame of results
total_summary_df = pd.DataFrame({"Number of Unique Items": [unique_items1],
                                "Average Price": [average_price],
                                "Number of Purchases": [number_purchase],
                                "Total Revenue": [total_revenue]
                                })
total_summary_df
```

```python
total_summary_df.dtypes
```

```python
# Converting avg price and total revenue to float
total_summary_df["Average Price"] = total_summary_df["Average Price"].astype(float)
total_summary_df["Total Revenue"] = total_summary_df["Total Revenue"].astype(float)
# Formatting to 2 decimal places and adding $
total_summary_df["Average Price"]=total_summary_df["Average Price"].map("${:.2f}".format)
total_summary_df["Total Revenue"]=total_summary_df["Total Revenue"].map("${:,.2f}".format)
total_summary_df
```

```python
# Count of male and female players, display in df, rename column
gender_count = purchase_data_df["Gender"].value_counts()
gender_count_df = pd.DataFrame(gender_count)
gender_count_df = gender_count_df.rename(
    columns={"Gender": "Total Count"})
gender_count_df
```

```python
# Percentage of male and female players, add column to df 
gender_total = gender_count_df["Total Count"].sum()
gender_percentage = gender_count_df["Total Count"] / gender_total * 100
gender_count_df["Percentage of Players"] = gender_percentage
gender_count_df.head()
```

```python
# Format percentage 
gender_count_df["Percentage of Players"]=gender_count_df["Percentage of Players"].map("{0:,.2f}%".format)
gender_count_df.head()
```

```python
# Grouping by gender
gender_group = purchase_data_df.groupby("Gender")

# Obtain purchase count, avg purchase price, total purchase value, avg purchase total per person
purchase_count = gender_group["Age"].count()
avg_purchase_price = gender_group["Price"].mean()
total_purchase_value = gender_group["Price"].sum()

# Create df 
purchase_gender_df = pd.DataFrame({"Purchase Count": purchase_count,
                                  "Average Purchase Price": avg_purchase_price,
                                  "Total Purchase Value": total_purchase_value,
                                  "Avg Purchase Total Per Person": total_purchase_value/gender_total,
                                  })
# formatting 
purchase_gender_df["Average Purchase Price"]=purchase_gender_df["Average Purchase Price"].map("${:.2f}".format)
purchase_gender_df["Total Purchase Value"]=purchase_gender_df["Total Purchase Value"].map("${:,.2f}".format)
purchase_gender_df["Avg Purchase Total Per Person"]=purchase_gender_df["Avg Purchase Total Per Person"].map("${:.2f}".format)

purchase_gender_df
```

```python
# Max and min ages
print(purchase_data_df["Age"].max())
print(purchase_data_df["Age"].min())
```

```python
# Establish bins
bins = [5, 10, 15, 20, 25,
        30, 35, 40, 45]

# Create labels for these bins
group_labels = ["<10", "10 to 14", "15 to 19", "20 to 24", "25 to 29", "30 to 34",
                "35 to 39", "40<"]

# Categorize 
purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"], bins, labels=group_labels)
purchase_data_df
```

```python
# Create a GroupBy object based upon "View Group"
age_group = purchase_data_df.groupby("Age Group")

# Calculate total count and percentage of players by age 
age_count = age_group["SN"].nunique()
total_player1 = age_count.sum()
percentage_player_age = (age_count / total_player1) * 100

# Create data frame
demographics_age = pd.DataFrame({"Total Counts": age_count,
                                "Percentage of Players": percentage_player_age})

demographics_age["Percentage of Players"]=demographics_age["Percentage of Players"].map("{0:,.2f}%".format)

demographics_age
```

```python
# Purchase count, avg purchase price, total purchase value, avg purchase total per person
purchase_count_age = age_group["Age"].count()
avg_purchase_age = age_group["Price"].mean()
total_purchase_age = age_group["Price"].sum()
avg_total_age = total_purchase_age / total_player1

# Create df 
purchase_age_df = pd.DataFrame({"Purchase Count": purchase_count_age,
                                  "Average Purchase Price": avg_purchase_age,
                                  "Total Purchase Value": total_purchase_age,
                                  "Avg Purchase Total Per Person": avg_total_age,
                                  })

purchase_age_df["Average Purchase Price"]=purchase_age_df["Average Purchase Price"].map("${:.2f}".format)
purchase_age_df["Total Purchase Value"]=purchase_age_df["Total Purchase Value"].map("${:.2f}".format)
purchase_age_df["Avg Purchase Total Per Person"]=purchase_age_df["Avg Purchase Total Per Person"].map("${:.2f}".format)

purchase_age_df
```

```python
# GroupBy based on SN
sn_group = purchase_data_df.groupby("SN")

# Purchase count, avg purchase price, total purchase value
purchase_count_sn = sn_group["Purchase ID"].count()
avg_purchase_sn = sn_group["Price"].mean()
total_purchase_sn = sn_group["Price"].sum()

# Create df
top_spender_df = pd.DataFrame({"Purchase Count": purchase_count_sn,
                                  "Average Purchase Price": avg_purchase_sn,
                                  "Total Purchase Value": total_purchase_sn,
                                  })

# Sort total purchase value in descending order 
top_spender_df = top_spender_df.sort_values("Total Purchase Value", ascending=False)

# format values, round decimal, add $
top_spender_df["Average Purchase Price"]=top_spender_df["Average Purchase Price"].map("${:.2f}".format)
top_spender_df["Total Purchase Value"]=top_spender_df["Total Purchase Value"].map("${:.2f}".format)
top_spender_df.head()
```

```python
# GroupBy based on Item ID and Item Name
item_group = purchase_data_df.groupby(['Item ID', 'Item Name'])

# Purchase count, avg purchase price, total purchase value
purchase_count_item = item_group["Purchase ID"].count()
avg_purchase_item = item_group["Price"].mean()
total_purchase_item = item_group["Price"].sum()

# Create df
popular_item_df = pd.DataFrame({"Purchase Count": purchase_count_item,
                                  "Average Purchase Price": avg_purchase_item,
                                  "Total Purchase Value": total_purchase_item,
                                  })

# Sort purchase count in descending order 
popular_item_df = popular_item_df.sort_values("Purchase Count", ascending=False)

# format values, round decimal, add $
popular_item_df["Average Purchase Price"]=popular_item_df["Average Purchase Price"].map("${:.2f}".format)
popular_item_df["Total Purchase Value"]=popular_item_df["Total Purchase Value"].map("${:.2f}".format)
popular_item_df
```
